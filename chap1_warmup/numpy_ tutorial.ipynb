#### 15. 利用13题目中的x，y 输出 x*y ,和 np.multiply(x, y) 还有  np.dot(x,y),比较差异。然后自己换一个不是方阵的试试。

```python
print('x:\n',x,'\n','y:\n',y)
print("x*y:\n",x*y)
print("np.multiply(x,y):\n",np.multiply(x,y))
print("np.dot(x,y):\n",np.dot(x,y))
print("本人的猜测，\n直接乘、multiply是按位乘，\ndot方法才是矩阵乘法.")
x1 = np.array([[1,1],[1,1]])
y1 = np.array([[8,4],[2,1]])
print("*,按位乘:\n",x1*y1)
print("np.multiply,按位乘:\n",np.multiply(x1,y1))
print("np.dot,矩阵乘:\n",np.dot(x1,y1))
```
